```python
#cargando librerias
import pandas as pd
import numpy as np
from datetime import datetime
from dateutil.parser import parse
import calendar as cal
import datetime as dat

# -*- coding: UTF-8 -*-

#definir el nombre archivo
filename ='DataSetNike.csv'

#cargar el archivo a un dataset
df = pd.read_csv(filename, header=0,encoding='utf-8',dtype={"Store":str,"Date":str,"No_Tran":int,"UPC":str,"Material":str,"Size":str,"Division":str,"Global Category":str,"Gender":str,"Public Price":float,"Units":int,"Sale Price Wo tax":float,"Tax":float,"Sale Price":float,"Discount":float,"Promo":str,"Sales Type ":str,"Emp No":str,"Sales Man":int,"TTL Cost":float})

#Se eliminaron los registros(filas) donde la variable Material estaba vacía. 228 registros.
df=df.dropna(subset=["Material"])

#Se eliminaron los registros donde la variable Units era menor a 1 . 5561 registros.
df=df[df["Units"]>=1]

#renombrar material a codigo
df.rename (columns ={"Material":"Codigo"}, inplace=True)

#dejar en el dataset solo las columnas requeridas
df = df.drop(['Store','No_Tran','UPC','Size','Division','Global Category','Gender','Public Price','Sale Price Wo tax','Tax','Sale Price','Discount','Promo','Sales Type ','Emp No','Sales Man','TTL Cost'], axis=1)

df['Date'] = pd.to_datetime(df["Date"])

#insertar columna para año
df.insert(2, "Year", df["Date"].apply(lambda x: x.year), True)

#insertar columna para mes
df.insert(3, "Mes", df["Date"].apply(lambda x: x.month) , True)

df.sort_values(by=["Year"], ascending=True, inplace=True)

#eliminar la columna Date
df=df.drop (["Date"], axis=1)

tabla = pd.pivot_table(df,values="Units", index=["Year","Codigo"], columns=["Mes"], aggfunc=np.sum, fill_value=0)

df = pd.DataFrame(tabla.to_records())

# definir los nuevos nombres de las columnas
newcols = {
    'Year': 'AÑO',
    'Codigo': 'CÓDIGO',
    '1': 'ENE.',
    '2': 'FEB.',
    '3': 'MAR.',
    '4': 'ABR.',
    '5': 'MAY.',
    '6': 'JUN.',
    '7': 'JUL.',
    '8': 'AGO.',
    '9': 'SEP.',
    '10': 'OCT.',
    '11': 'NOV.',
    '12': 'DIC.'
}

# renombrar columnas
df.rename(columns=newcols, inplace=True)

#cambiar el orden de las columnas
df = df.reindex(columns=["CÓDIGO","AÑO","ENE.","FEB.","MAR.","ABR.","MAY.","JUN.","JUL.","AGO.","SEP.","OCT.","NOV.","DIC."])

#asignar un nombre al nuevo archivo
outputfile="DataSetNike_sku.csv"

# saving the dataframe
df.to_csv(outputfile, header=True, index=False, encoding='utf-8-sig') 
```
